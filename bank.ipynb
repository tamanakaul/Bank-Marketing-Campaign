```python
import matplotlib.pyplot as plt

df['job'].value_counts().plot(
    kind='bar',
    figsize=(10,5),
    color='skyblue',
    edgecolor='black'
)

plt.title("Customer Distribution by Job")
plt.xlabel("Job")
plt.ylabel("Number of Customers")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
df['marital'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Marital Status Distribution")
plt.ylabel("")
plt.show()
```

```python
df['education'].value_counts().plot(
    kind='bar',
    color='orange'
)

plt.title("Education Level")
plt.xlabel("Education")
plt.ylabel("Customers")
plt.show()
```

```python
df['housing'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Housing Loan")
plt.ylabel("")
plt.show()
```

```python
df['loan'].value_counts().plot(
    kind='bar',
    color='green'
)

plt.title("Personal Loan Status")
plt.xlabel("Loan")
plt.ylabel("Count")
plt.show()
```

```python
df['age'].plot(
    kind='hist',
    bins=20,
    color='purple'
)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
df['contact'].value_counts().plot(
    kind='bar',
    color='red'
)

plt.title("Contact Method")
plt.xlabel("Contact Type")
plt.ylabel("Customers")
plt.show()
```

```python
df['month'].value_counts().plot(
    kind='barh',
    color='teal'
)

plt.title("Campaign by Month")
plt.xlabel("Number of Customers")
plt.ylabel("Month")
plt.show()
```

```python
df['job'].value_counts().head(10).plot(
    kind='barh',
    color='gold'
)

plt.title("Top 10 Customer Jobs")
plt.xlabel("Count")
plt.show()
```

```python
df['balance'].plot(
    kind='hist',
    bins=30,
    color='cyan'
)

plt.title("Account Balance Distribution")
plt.xlabel("Balance")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df['age'],
    df['balance'],
    alpha=0.5,
    color='blue'
)

plt.title("Age vs Account Balance")
plt.xlabel("Age")
plt.ylabel("Balance")
plt.show()
```

```python
df['poutcome'].value_counts().plot(
    kind='bar',
    color='magenta'
)

plt.title("Previous Campaign Outcome")
plt.xlabel("Outcome")
plt.ylabel("Count")
plt.show()
```
